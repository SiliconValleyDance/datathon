# 대출 정보랑 join 해서 보기

```python
rental_data = pd.read_csv("대출정보.csv", encoding="latin1")
```

```python
rental_data.columns = ["도서ID", "대출일자"]
```

```python
rental_data.head()
```

```python
# 1. 대출일자를 datetime 형식으로 변환
rental_data["대출일자"] = pd.to_datetime(rental_data["대출일자"])

# 2. 대출일자를 기준으로 연도-월별 대출 건수 집계
rental_data["연도-월"] = rental_data["대출일자"].dt.to_period("M")  # 연도-월 단위로 변환
monthly_counts = rental_data.groupby("연도-월").size()  # 그룹화하여 개수 계산

# 3. 시각화
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("연도-월별 도서 대출 건수", fontsize=16)
plt.xlabel("연도-월", fontsize=14)
plt.ylabel("대출 건수", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2022년 데이터만 필터링
monthly_counts_2022 = monthly_counts[monthly_counts.index.year == 2022]

# 시각화
plt.figure(figsize=(10, 6))
monthly_counts_2022.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("2022년 연도-월별 도서 대출 건수", fontsize=16)
plt.xlabel("연도-월", fontsize=14)
plt.ylabel("대출 건수", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2022년 데이터만 필터링
monthly_counts_2021 = monthly_counts[monthly_counts.index.year == 2021]

# 시각화
plt.figure(figsize=(10, 6))
monthly_counts_2021.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("2021년 연도-월별 도서 대출 건수", fontsize=16)
plt.xlabel("연도-월", fontsize=14)
plt.ylabel("대출 건수", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2021년과 2022년 데이터 필터링
monthly_counts_2020 = monthly_counts[monthly_counts.index.year == 2020]
monthly_counts_2021 = monthly_counts[monthly_counts.index.year == 2021]
monthly_counts_2022 = monthly_counts[monthly_counts.index.year == 2022]

# 두 데이터를 하나의 DataFrame으로 합치기
comparison_data = pd.DataFrame({
    "2020" : monthly_counts_2020.values,
    "2021": monthly_counts_2021.values,
    "2022": monthly_counts_2022.values
}, index=monthly_counts_2021.index.strftime("%Y-%m"))  # 연도-월을 문자열로 변환하여 인덱스로 설정

# 시각화
plt.figure(figsize=(12, 6))
comparison_data.plot(kind="bar", color=["skyblue", "orange", "green"], edgecolor="black")
plt.title("2020 - 2022년 연도-월별 도서 대출 건수 비교", fontsize=16)
plt.xlabel("월", fontsize=14)
plt.ylabel("대출 건수", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.legend(title="연도", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2013년 데이터만 필터링
monthly_counts_2013 = monthly_counts[monthly_counts.index.year == 2013]

# 시각화
plt.figure(figsize=(10, 6))
monthly_counts_2013.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("2013년 연도-월별 도서 대출 건수", fontsize=16)
plt.xlabel("연도-월", fontsize=14)
plt.ylabel("대출 건수", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
