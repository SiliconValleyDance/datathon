```python
old_rentals["대출연도"] = pd.to_datetime(old_rentals["대출일자"]).dt.year
book_rental_info_in_old = old_rentals.groupby(["도서ID", "대출연도"]).size().unstack(fill_value=0)
book_rental_info_in_old
```
