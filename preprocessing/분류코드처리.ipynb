```python

for n in range (0, len(books)) :
    try :
        kdc_code = "{:03}".format(int(re.search(r'\d+', books.loc[n, '분류코드']).group()))
        #"{:03}".format(int(re.split('[.,A-Za-z\s]', books.loc[n, '분류코드'])[0]))
    except ValueError :
        print(n, f"kdc_code format err{books.loc[n, '분류코드']}")
        if books.loc[n, "분류코드"] is None :
            books.loc[n, "주류"] = "기타"
            books.loc[n, "강목"] = "기타"
            books.loc[n, "요목"] = "기타"
            
            print("processing None")
            
        continue
        
    try : 
        first = int(kdc_code[0]) * 100
        second = first + int(kdc_code[1]) * 10
        third = second + int(kdc_code[2])

        first = f"{first:03}"
        second = f"{second:03}"
        third = f"{third:03}"

        books.loc[n, "주류"] = main_code[first]
        books.loc[n, "강목"] = gang_code[second]
        books.loc[n, "요목"] = yo_code[third]
        
    except ValueError :
        print(n, kdc_code)
        continue
```
